<h1>MonteCarlo</h1>


```python
import matplotlib.pyplot as plt
import numpy as np

n = 160
numExperimentos = 50
mediaPi = 0
listaValoresPi = []
for i in range(numExperimentos):
    valor = 0
    X = np.random.uniform(0, 1, n).tolist()
    Y = np.random.uniform(0, 1, n).tolist()
    for j in range(n):
        Z = np.sqrt(X[j] * X[j] + Y[j] * Y[j])
        if Z <= 1:
            #Estamo dentro del cÃ­rculo
            valor += 1
    valorFlotante = float(valor)
    valorPi = valorFlotante * 4 / n
    listaValoresPi.append(valorPi)
    mediaPi += valorPi

    
piFinal = mediaPi / numExperimentos
print("Resultado ", piFinal)
```

```python
def mc_pi_aprox(N=10000):
    plt.figure(figsize=(8,8))  # tamaÃ±o de la figura
    x, y = np.random.uniform(-1, 1, size=(2, N))
    interior = (x**2 + y**2) <= 1
    pi = interior.sum() * 4 / N
    error = abs((pi - np.pi) / pi) * 100
    exterior = np.invert(interior)
    plt.plot(x[interior], y[interior], 'b.')
    plt.plot(x[exterior], y[exterior], 'r.')
    plt.plot(0, 0, label='$\hat \pi$ = {:4.4f}\nerror = {:4.4f}%'
             .format(pi,error), alpha=0)
    plt.axis('square')
    plt.legend(frameon=True, framealpha=0.9, fontsize=16)

mc_pi_aprox()
```
