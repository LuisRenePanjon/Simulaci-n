```python
import random
import matplotlib.pyplot as plot
def run_dices(n):
	arr_sum = []
	for i in range(n):
		dice1 = random.randint(1, 6)
		dice2 = random.randint(1, 6)
		arr_sum.append(dice1 + dice2)

	values = range(1, 15)
	plot.hist(x=arr_sum, bins=values , color='#fe6b36', rwidth=0.85)
	plot.title('Histograma de Sum de dados')
	plot.xlabel('Sumatoria')
	plot.ylabel('Frecuencia')
	plot.xticks(values)

	plot.show()
```

<h1>Resultados</h1>

```python
run_dices(100)
```

```python
run_dices(1000)
```

```python
run_dices(10000)
```
